# Regresión Logística

La regresión logística es la manera estándar de modelar variables respuesta binarias.

## Predictores únicos

Para ilustar el funcionamiento básico de una regresión logística, se empieza partiendo de una regresión con un único predictor de un problema de preferencias políticas electorales en las elecciones de 1992 de USA, ignorando aquellos votos que se abstuvieron o que votaron por representantes distintos a los partidos Republicano y Demócrata

```python
# Carga de datos completos
datos = pd.read_csv(
    os.path.join(DATA, 'ideo.csv'))

# Filtrar datos para regresión pred. único
datos_predictor_unico = (
    datos
    .query(expr="year == 1992")
    .query(expr="presvote < 3")
    .filter(items=['income', 'presvote'])
    .assign(vote = lambda _df: _df.presvote - 1)
    .reset_index(drop=True))
```

Se hace una inspección visual rápida de los datos

```python
# Especificaciones de figura
fig, ax = plt.subplots(figsize=(8,5))

# Elementos de visualización
x_jitter = np.random.uniform(low=-0.25, high=0.25, size=datos_predictor_unico.shape[0])
y_jitter = np.random.uniform(low=-0.05, high=0.05, size=datos_predictor_unico.shape[0])

ax.scatter(
    x=datos_predictor_unico.income + x_jitter
    ,y=datos_predictor_unico.vote + y_jitter
    ,s=10
    ,color=CONFIG['graficas']['colors']['azul']
    ,alpha=0.5)

# Anotaciones y estilo
ax.set(
    title='Afinidad política en las encuestas'
    ,xlabel='Categoría de ingreso'
    ,ylabel='Vota por P. Republicano')

plt.show()
```

Al obsevar los datos, en realidad se puede ver que no hace ningún sentido modelar una regresión lineal ya que esto excedería el dominio de la variable predictora. En cambio, se puede modelar la probabilidad de que $y=1$ con la expresión:

$$P(y=1)=logit^{-1}(X_i\beta)$$

Bajo el supuesto de que las respuestas $y_i$ son independientes dada la distribución de probabilidad. El término $X_i\beta$ representa los predictores lineales, y la función $logit^{-1}$ se encarga de transformar valores continuos al rango $(0, 1)$, lo cual es necesario dado que se están modelando probabilidades, y está dado por la expresión:

$$logit^{-1}(x)=\frac{e^x}{1+e^x}$$

Para entenderlo mejor, otra forma de ver el modelo es la siguiente. Se sabe que la probabilidad $P(y_i=1)\in(0,1)$, por lo que si se iguala a un valor $p_i$ entonces $p_i\in(0,1)$, es decir:

$$P(y_i=1)=p_i$$

Esta probabilidad proviene de las variables explicativas $X_i\beta\in\mathbb{R}$, y sabiendo que la función $logit(x): x\in(0,1)\rightarrow x\in\mathbb{R}$, entonces se puede justificar la igualdad:

$$logit(p_i)=X_i\beta$$

Que al aplicar la función $logit^{-1}$ en ambos lados de la igualdad, recurpera la expresión original.

### Modelo con `sklearn`

```python
# Datos de modelado
x_train = datos_predictor_unico.income.to_numpy().reshape((-1,1))
y_train = datos_predictor_unico.vote.to_numpy()

# Iniciar instancia de modelo
log_reg = LogisticRegression()

# Entrenamiento y resumen
results = log_reg.fit(x_train, y_train)
results.intercept_, results.coef_
```

### Modelo con `statsmodels`

```python
# Datos de modelado
x_train = datos_predictor_unico.income
y_train = datos_predictor_unico.vote

# Agregar intercepto
x_train = sm.add_constant(data=x_train)

# Iniciar instancia de modelo
log_reg = sm.Logit(endog=y_train, exog=x_train)

# Entrenamiento y resumen
results = log_reg.fit()
results.summary()
```

### Visualización del modelo

Los resultados del modelo anterior pueden visualizarse si se resuelve el modelo para un rango dado

```python
# Malla para mapear el modelo
x_vis = np.linspace(start=-5, stop=10, num=datos_predictor_unico.shape[0])

# Predicciones del modelo
y_vis = np.array([
    expit(results.params['const'] + results.params['income'] * x_vis)
    ,log_reg.cdf(X=x_vis)
])

# Especificaciones de figura
fig, ax = plt.subplots(figsize=(8,5))

# Elementos de visualización
ax.plot(
    x_vis, y_vis[0]
    ,color=CONFIG['graficas']['colors']['verde']
    ,label='model')
ax.plot(
    x_vis, y_vis[1]
    ,color=CONFIG['graficas']['colors']['amarillo']
    ,label='invLogit')
ax.scatter(
    x=datos_predictor_unico.income + x_jitter
    ,y=datos_predictor_unico.vote + y_jitter
    ,s=10
    ,label='data'
    ,color=CONFIG['graficas']['colors']['azul']
    ,alpha=0.5)

# Anotaciones y estilo
ax.set(
    title='Regresión logística con predictor único'
    ,xlabel='Categoría de ingreso'
    ,ylabel='Probabilidad de voto por P. Republicano')
ax.legend()

plt.show()
```

Para interpretar estos resultados es necesario evaluar el cambio en la probabilidad $P(y_i=1)$ cuando $x$ toma algún valor, por ejemplo la media $\bar{x}$

```python
# Promeido de categoría de ingreso
x_mean = datos_predictor_unico.income.mean()

print(f"P(y=1) en x=media: {expit(results.params['const'] + results.params['income'] * x_mean) * 100:.2f} %")
```

En el valor promedio de la categoría de ingreso, la $P(y=1)=40\%$.

Otra forma de interpretar los coeficientes es evaluando el cambio en la probabilidad cuando $x$ cambia en una unidad, por ejemplo, de una categoría de ingreso $2\rightarrow 3$

```python
# Probabilidades que se van a comparar
p_2 = expit(results.params['const'] + results.params['income'] * 2)
p_3 = expit(results.params['const'] + results.params['income'] * 3)

print(f"Cambio en P(y=1) cuando X: (2 -> 3): {(p_3 - p_2) * 100:.2f} %")
```

Pasar de un nivel de ingreso de $2\rightarrow 3$ incrementa la probabilidad $P(y=1)$ en $7.5\%$

### Visualizando múltiples modelos

En este caso, dado que se trata de un modelo con solamente una variable explicativa, sería posible hacer una visualización en la que se muestre la evolución del coeficiente a lo largo de los años, para lo cual es necesario ajustar un modelo por cada año en el cual haya registros

```python
# Función para generar modelo condicionado al año de análisis
def modelo_por_anio(year):
    try:
        _df = (
            datos
            .query(expr=f"year == {year}")
            .query(expr="presvote < 3")
            .filter(items=['income', 'presvote'])
            .assign(vote = lambda _df: _df.presvote - 1)
            .reset_index(drop=True))
    
        # Datos de modelado
        x_train = _df.income
        y_train = _df.vote
        
        # Agregar intercepto
        x_train = sm.add_constant(data=x_train)
        
        # Iniciar instancia de modelo
        log_reg = sm.Logit(endog=y_train, exog=x_train)
        
        # Entrenamiento y resumen
        results = log_reg.fit(disp=False)
        results = {
            'income':results.params.income
            ,'std_err': results.bse.income}

    except ValueError:
        results = np.nan

    return results

# Modelado para múltiples años
multiples_modelos = {year: modelo_por_anio(year) for year in datos.year.unique()}
multiples_modelos = pd.DataFrame(data=multiples_modelos).T.dropna()

multiples_modelos.head(3)
```

Con la tabla anterior se puede construir la siguiente visualización:

```python
# Especificaciones de figura
fig, ax = plt.subplots(figsize=(8,5))

# Elementos de visualización
# ---- Coeficiente
ax.scatter(
    x=multiples_modelos.index
    ,y=multiples_modelos.income
    ,s=15
    ,color=CONFIG['graficas']['colors']['rojo'])

# ---- Intervalo a +- 1 desv. estándar
for row in multiples_modelos.index:
    ax.vlines(
        x=row
        ,ymin=multiples_modelos.loc[row].income - multiples_modelos.loc[row].std_err
        ,ymax=multiples_modelos.loc[row].income + multiples_modelos.loc[row].std_err
        ,colors=CONFIG['graficas']['colors']['rojo']
        ,linewidth=1)

# Anotaciones y estilo
ax.set(
    title='Efecto del ingreso en afinidad política a lo largo del tiempo'
    ,xlabel='Año'
    ,ylabel='Coeficiente de ingreso')

plt.show()
```

El efecto que puede observarse es que, en general, la relevancia del ingreso ha ido aumentando a lo largo de los años para este modelo en particular.

## Ejemplo completo: Pozos en Bangladesh

A continuación se ilustran los pasos que se deben seguir en un problema de regresión logística, en donde el problema fundamental consiste en modelar la decisión de dueños de tierras en Bangladesh sobre si cambiar o no su suministro de agua potable.

### Contexto del problema

Muchos de los pozos de suministro de agua potable en Bangladesh y otras sonas del sur de Asia están contaminados con arsénico natural. El arsénico es un veneno acumulativo cuya exposición aumenta el riesgo de cáncer y otras enfermedades. Cualquier localidad puede incluir pozos con diversos niveles de arsénico, y el que un pozo sea seguro no quiere decir que otro pozo contiguo también lo sea. 

Un equipo de investigadores midieron todos los pozos de la zona en cientos de microgramos de arsénico por litro de agua, y los etiquetaron de la siguiente manera:

* _safe_: Si los niveles de arsénico son menores a 0.5.
* _unsafe_: Si los niveles de arsénico son mayores a 0.5.

A las personas que tenían acceso a pozos inseguros se les exhortó a cambiar a una zona cercana con un pozo seguro. Tiempo después, los investigadores regresaron para determinar quiénes siguieron las recomendaciones y cambiaron de pozo.

El **objetivo** es entonces realizar un análisis de regresión logística  para entender los factores predictivos del cambio de pozos:

* $1$: Si el dueño $i$ cambió a un nuevo pozo.
* $0$: Si el dueño $i$ no cambió de pozo

Las entradas consideradas son las siguientes:

* Un término constante
* La distancia al pozo seguro más cercano
* Nivel de arsénico en el pozo de los dueños
* Indicador de si alguno de los dueños del pozo es un miembro activo en organizaciones de su comunidad
* Nivel educativo

### Un primero modelo

El primer modelo en el que se puede pensar es en aquel que relacione a la variable respuesta con la distancia que hay hacia el pozo seguro más cercano, en cuyo caso:

$$P(switch=1) = p_i$$
$$logit(p_i) = \alpha + \beta\cdot dist$$

Podemos caracterizar rápidamente la distribución de la distancia al pozo seguro más cercano con un histograma

```python
# Carga de datos completos
datos = pd.read_csv(
    os.path.join(DATA, 'wells.csv'))

datos.info()
```

```python
# Especificaciones de figura
fig, ax = plt.subplots(figsize=(8,5))

# Elementos de visualización
ax.hist(
    x=datos.dist
    ,bins=30
    ,color=CONFIG['graficas']['colors']['azul'])

# Anotaciones y estilo
ax.set(
    xlabel='Distancia al pozo seguro más cercano [m]')

plt.show()
```

Se hace el primero modelo utilizando funciones de `statsmodels`

```python
# Datos de modelado
x_train = datos.dist
y_train = datos.switch

# Agregar intercepto
x_train = sm.add_constant(data=x_train)

# Iniciar instancia de modelo
log_reg = sm.Logit(endog=y_train, exog=x_train)

# Entrenamiento y resumen
results = log_reg.fit()
results.summary()
```
